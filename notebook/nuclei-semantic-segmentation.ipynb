```python
import cv2,os
```

```python
# Initialize empty lists for images and masks from the dataset
images = []
masks = []
PATH = r"../datasets/train"
```

```python
# Load images from datasets using openCV
IMG_PATH = os.path.join(PATH,'inputs')
for image_file in os.listdir(IMG_PATH):
    img = cv2.imread(os.path.join(IMG_PATH,image_file))
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    img = cv2.resize(img,(128,128))
    images.append(img)
```

```python
MASKS_PATH = os.path.join(PATH, 'masks')
for mask_file in os.listdir(MASKS_PATH):
    mask = cv2.imread(os.path.join(MASKS_PATH, mask_file))
    mask = cv2.resize(mask,(128,128))
    masks.append(mask)
```

```python
import numpy as np
```

```python
# Convert lists into numpy array
images_np = np.array(images)
masks_np = np.array(masks)
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# display images example from the dataset
plt.figure(figsize=(10,10))
for i in range(1,4):
    plt.subplot(1,3,i)
    plt.imshow(images_np[i])
    plt.axis('off')
    
plt.show()
```

```python
# display images example from the dataset
plt.figure(figsize=(10,10))
for i in range(1,4):
    plt.subplot(1,3,i)
    plt.imshow(masks_np[i])
    plt.axis('off')

plt.show
```
